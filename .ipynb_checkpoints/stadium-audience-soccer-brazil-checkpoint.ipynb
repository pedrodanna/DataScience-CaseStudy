# Data Science Case Study (Stadium Audiance in Brazil)

In this case study we're going to discover what most impacts the audience at the stadium of Brazilian soccer league first division.

![title](img/torcida_brasil.png)

## Python Libraires Imports

```python
import pandas as pd
```

## General Yearly Average

I'm going to get the Data Frame of the general yearly average of the stadium audience among the years.

```python
#Extraction and cleaning of the data from the source

#Getting the desired table from the web page
gya = pd.read_html('https://pt.wikipedia.org/wiki/P%C3%BAblicos_no_Campeonato_Brasileiro_de_Futebol')[3]
#Initializing in the year 1969 because of league regulations
gya = gya[12:]
#Translating columns names to English
gya.columns = ['Year', 'Audience']
#Cleaning Year column dirt
gya['Year'] = gya['Year'].apply(lambda x: x[:4])
#Transforming Audience column to number
gya['Audience'] = gya['Audience'].apply(lambda x: int(x.replace('.', '')))
#Setting Year column as index
gya.set_index('Year', inplace=True)
gya
```

Now I'm going to plot this Data Frame into a histogram.<br>
(<b>Double click the graph for a better visualization</b>)

```python
ytk = [0, 5000, 10000, 15000, 20000, 25000]
ax = gya.plot(kind='bar',figsize=(51,10), color='g', yticks=ytk)
for l in ax.patches:
    ax.annotate(str(l.get_height()), (l.get_x(), l.get_height() * 1.030))
```
